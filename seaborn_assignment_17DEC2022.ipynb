```python
import numpy as np
import pandas as pd

gradebook = pd.read_csv('https://raw.githubusercontent.com/EnggQasim/PGD_UGD_Data_Visualization/main/class14_matplotlib_seaboarn/gradebook.csv')

print(gradebook)
```

```python
# 3. Select all rows from the gradebook DataFrame where assignment_name is equal to Assignment 1. 
#Save the result to the variable assignment1. Print assignment1.

assignment1 = gradebook.loc[gradebook['assignment_name'].str.startswith('Assignment 1')]
print(assignment1)
```

```python
#Now use Numpy to calculate the median grade in assignment1. 
#Use np.median() to calculate the median of the column grade from assignment1 and save it to asn1_median.

asn1_median= np.median(assignment1['grade'])
print('Median Grade on Assignment: ', asn1_median)
```

data equal to df

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df1 = pd.read_csv('https://raw.githubusercontent.com/EnggQasim/PGD_UGD_Data_Visualization/main/class14_matplotlib_seaboarn/survey.csv')

sns.barplot(data=df1, x='Gender', y='Response', estimator=len)
plt.show()
```

```python
#3. Change sns.barplot() to graph the median Response aggregated by Gender using estimator=np.median.
sns.barplot(data=df1, x='Gender', y='Response', estimator=np.median)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
n = 500
set1="https://raw.githubusercontent.com/EnggQasim/PGD_UGD_Data_Visualization/main/class14_matplotlib_seaboarn/dataset1.csv"
set2="https://raw.githubusercontent.com/EnggQasim/PGD_UGD_Data_Visualization/main/class14_matplotlib_seaboarn/dataset2.csv"
set3="https://raw.githubusercontent.com/EnggQasim/PGD_UGD_Data_Visualization/main/class14_matplotlib_seaboarn/dataset3.csv"
dataset1 = np.genfromtxt(set1, delimiter=",")
dataset2 = np.genfromtxt(set2, delimiter=",")
dataset3 = np.genfromtxt(set3, delimiter=",")
df = pd.DataFrame({
    "label": ["set_one"] * n + ["set_two"] * n + ["set_three"] * n,
    "value": np.concatenate([dataset1, dataset2, dataset3])
})

sns.set()
sns.barplot(data=df, x='label', y='value')
plt.show()
```

```python
sns.kdeplot(dataset1, shade=True)
sns.kdeplot(dataset2, shade=True)
sns.kdeplot(dataset3, shade=True)
plt.legend()
plt.show()
```

```python
sns.kdeplot(dataset1, shade=True, label="dataset1")
sns.kdeplot(dataset2, shade=True, label="dataset2")
sns.kdeplot(dataset3, shade=True, label="dataset3")

plt.legend()
plt.show()
```

```python
sns.boxplot(data=df, x='label', y='value')
plt.show()
```

```python
sns.violinplot(data=df, x="label", y="value")
plt.show()
```
