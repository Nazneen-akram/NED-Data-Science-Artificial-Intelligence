```python
# Load the image-to-text pipeline with the specified model
captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
# Provide the URL of the image you want to caption
image_url = "https://huggingface.co/datasets/Narsil/image_dummy/resolve/main/parrots.png"
```

![parrots.png](lena.png)

```python
# Generate captions for the image
captions = captioner(image_url)
```

```python
# Print the generated captions
print(captions)  # Output: [{'generated_text': 'two birds are standing next to each other '}]
```

```python
model_image = 'https://huggingface.co/datasets/Narsil/image_dummy/raw/main/lena.png'
```

```python
# Generate captions for the image
captions = captioner(model_image)
```

```python
# Print the generated captions
print(captions)  # Output: [{'generated_text': 'two birds are standing next to each other '}]
```
