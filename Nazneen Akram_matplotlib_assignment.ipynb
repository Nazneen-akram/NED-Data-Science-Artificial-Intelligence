```python
from matplotlib import pyplot as plt
time = [0,1,2,3,4]
revenue = [200,400,650,800,850]
plt.plot(time,revenue)
plt.title('Revenue over Time (in years)')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
time = [0,1,2,3,4]
revenue = [200,400,650,800,850]
costs = [150,500,550,550,560]

plt.plot(time,revenue, marker ='o')
plt.plot(time,costs, marker='x')

plt.title('Trend of Revene Vs. Costs over time (in years)')
plt.xlabel('Time')
plt.ylabel('Revenue / Costs ($)')
plt.show()
```

```python
time = [0,1,2,3,4]
revenue = [200,400,650,800,850]
costs = [150,500,550,550,560]

plt.plot(time,revenue, color = 'purple', linestyle='--', marker ='o')
plt.plot(time,costs, color='#82edc9', marker='x')

plt.title('Trend of Revene Vs. Costs over time (in years)')
plt.xlabel('Time')
plt.ylabel('Revenue / Costs ($)')
plt.show()
```

```python
import numpy as np

drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

store1_x = np.arange(len(drinks))

plt.bar(store1_x - 0.2,sales1,0.4,label='Sales1')
plt.bar(store1_x + 0.2,sales2,0.4,label='Sales2')

plt.xticks(store1_x, drinks)
plt.legend()
plt.show()
```

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

plt.bar(range(len(sales1)),sales1)
plt.bar(range(len(sales2)),sales2, bottom=sales1)
plt.xticks(range(len(drinks)), drinks)
plt.legend(['Location 1', 'Location 2'], loc=1)
plt.show()
```
